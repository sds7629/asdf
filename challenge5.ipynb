문제1  

print_even 리스트 내에 있는 값 중 짝수만 화면에 출력되도록 print_score 함수를 생성 후 결과값을 출력해주세요

print_even ([1, 3, 2, 10, 12, 11, 15])

```python
new_list = [1,3,2,10,12,11,15]
def print_even(numbers: list) -> list:
    even_list = []
    for i in range(len(numbers)):
        if numbers[i] % 2 == 0:
            even_list.append(numbers[i])
    return even_list
print_even(new_list)
```

문제5  

오즈 (oz) 클래스에 (이름, 나이, 코딩레벨)을 받는 생성자를 추가해주세요

coding = oz("김코", 30, "A")


```python
class oz_info:
    def __init__(self, name:str = "김조교", age:int = 10, level:str = "F") -> None:
        self.name = name
        self.age = age
        self.level = level
    def my_info(self) -> None:
        print(f"이름은 {self.name}이고, 나이는 {self.age}이며, 코딩레벨은 {self.level}이야")

coding2 = oz_info("김코",30, "A")
print(coding2.name)
print(coding2.age)
print(coding2.level)
```
