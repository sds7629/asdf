```python
import random

words = ["apple", "banana", "melon", "dragonfruit", "cherry"]
your_answer = []

def make_word(words):
    word = random.choice(words)
    return word

def hang():
    choice_word = make_word(words)
    global your_answer
    for _ in range(len(choice_word)):
        your_answer += "-"
    
    your_life = len(set(choice_word))

    while True:
        print(your_answer)
        your_guess = input("ì¶”ì¸¡í•œ ë‹¨ì–´: ")


        for idx, letter in enumerate(choice_word):
            if letter == your_guess:
                your_answer[idx] = letter

        if your_guess not in choice_word:
            your_life -= 1
            if your_life == 0:
                print("gameover")
                break
        if "-" not in your_answer:
            print("you win")
            print("".join(your_answer))
            break

        
        print(f"{your_life}")

hang()
```
